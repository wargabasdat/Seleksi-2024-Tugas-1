```python
import pandas as pd
import os
import json
from datetime import datetime
```

```python
# Get a list of all JSON files in the folder
json_folder_path = "../../Data Scraping/data"
json_files = [f for f in os.listdir(json_folder_path) if f.endswith('.json')]

json_files
```

```python
with open(json_folder_path+"/"+"anime_list.json", 'r') as f:
    anime_dict = json.load(f)
    
with open(json_folder_path+"/"+"clists_list.json", "r") as f:
    clists_dict = json.load(f)

with open(json_folder_path+"/"+"hated_characters.json", "r") as f:
    char_dict = json.load(f)

with open(json_folder_path+"/"+"review_list.json", "r") as f:
    review_dict = json.load(f)

with open(json_folder_path+"/"+"users.json", "r") as f:
    users_dict = json.load(f)
```

```python
anime_in_list = []

for anime in anime_dict:
    for clists in clists_dict:
        if clists["list_title"] in anime["contained_in"]:
            anime_in_list.append({"username": clists["creator"],
                                  "list_title": clists["list_title"],
                                  "anime_title": anime["title"],
                                  "anime_year": anime["year"]})
    del anime["contained_in"]

anime_in_list
```

```python
appears_in = []

for char in char_dict:
    for anime in anime_dict:
        if anime["title"] in char["appears_in"]:
            appears_in.append({"char_name": char["name"],
                                  "anime_title": anime["title"],
                                  "anime_year": anime["year"]})
    del char["appears_in"]

appears_in
```

```python
for review in review_dict:
    for anime in anime_dict:
        if anime["title"] in review["anime_title"]:
            review["anime_year"] = anime["year"]
```

```python
character_traits = []
character_tags = []

for char in char_dict:
    for trait in char["traits"]:
        character_traits.append({"name": char["name"],
                                  "trait": trait})
    for tag in char["tags"]:
        character_tags.append({"name": char["name"],
                                  "tag": tag})
    del char["traits"]
    del char["tags"]
```

```python
with open('./intermediate/Animes.json', 'w') as outfile:
        json.dump(anime_dict, outfile, indent=4)
with open('./intermediate/Characters.json', 'w') as outfile:
        json.dump(char_dict, outfile, indent=4)
with open('./intermediate/Custom_Lists.json', 'w') as outfile:
        json.dump(clists_dict, outfile, indent=4)
with open('./intermediate/Reviews.json', 'w') as outfile:
        json.dump(review_dict, outfile, indent=4)
with open('./intermediate/Users.json', 'w') as outfile:
        json.dump(users_dict, outfile, indent=4)
with open('./intermediate/Animes_in_Lists.json', 'w') as outfile:
        json.dump(anime_in_list, outfile, indent=4)
with open('./intermediate/Appears_in.json', 'w') as outfile:
        json.dump(appears_in, outfile, indent=4)
with open('./intermediate/Characters_Traits.json', 'w') as outfile:
        json.dump(character_traits, outfile, indent=4)
with open('./intermediate/Characters_Tags.json', 'w') as outfile:
        json.dump(character_tags, outfile, indent=4)
```

```python
json_files = [f for f in os.listdir("./intermediate") if f.endswith('.json')]
f = open("./intermediate/sql_statements.txt", "w", encoding="utf-8")

sql_statements = []
    
for json_file in json_files:
    
    df = pd.read_json("./intermediate/"+json_file)
    
    for _, row in df.iterrows():
        # Escape single quotes in the values
        values = [f"'{str(value).replace("'", "''")}'" if isinstance(value, str)
                  else str(value) if isinstance(value, (int, float))
                  else f"'{value.strftime('%Y-%m-%d')}'" if isinstance(value, datetime)
                  else 'NULL' for value in row]
        # Join the column names and values to form the INSERT statement
        sql = f"INSERT INTO {json_file.split(".")[0]} ({', '.join(df.columns)}) VALUES ({', '.join(values)});"
        sql.replace("nan", "None")
        
        f.write(sql+"\n")
        sql_statements.append(sql)

f.close()
sql_statements
```
