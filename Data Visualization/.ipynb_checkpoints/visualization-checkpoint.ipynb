# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.merge(player, player_info, left_on='Player', right_on='Name')
```

```python
avg_points_by_position = df.groupby('Position')['Points per game'].mean().reset_index()

avg_points_by_position = avg_points_by_position.sort_values(by='Points per game', ascending=False)

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Position', y='Points per game', data=avg_points_by_position)
plt.title('Average Points per Game by Player Position')
plt.xlabel('Position')
plt.ylabel('Average Points per Game')



plt.show()
```

```python
position_counts = df['Position'].value_counts()
total = position_counts.sum()

position_percentages = (position_counts / total) * 100

above_threshold = position_percentages[position_percentages >= 2]
below_threshold = position_percentages[position_percentages < 2]

above_threshold['Other'] = below_threshold.sum()

plt.figure(figsize=(8, 8))
plt.pie(above_threshold, labels=above_threshold.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Player Positions')
plt.axis('equal')  
plt.show()
```

```python
# Plot histogram with KDE
plt.figure(figsize=(10, 6))
histplot = sns.histplot(df['3 Points Percentage'], bins=15, kde=True)

# Add the values on top of the bars
for patch in histplot.patches:
    height = patch.get_height()
    if height > 0:
        plt.text(
            patch.get_x() + patch.get_width() / 2,
            height,
            f'{int(height)}',
            ha='center',
            va='bottom'
        )

plt.title('Distribution of 3 Points Percentage')
plt.xlabel('3 Points Percentage')
plt.ylabel('Frequency')
plt.show()
```

```python
match['Date'] = pd.to_datetime(match['Date'])

match['Average Score'] = (match['Home Score'] + match['Away Score']) / 2
average_scores = match.groupby(match['Date'].dt.date)['Average Score'].mean().reset_index()


plt.figure(figsize=(12, 6))
plt.plot(average_scores['Date'], average_scores['Average Score'], marker='o',color='blue')
plt.xlabel('Date')
plt.ylabel('Average Score')
plt.title('Average Scores Over Time')
plt.grid(True)

for i in range(len(average_scores)):
    plt.annotate(f'{average_scores["Average Score"][i]:.1f}', 
                 (average_scores['Date'][i], average_scores['Average Score'][i]), 
                 textcoords="offset points", 
                 xytext=(0,10), 
                 ha='center',
                 fontsize=6, 
                 color='red') 

plt.show()
```

```python
match['Winner'] = match.apply(lambda x: 'Home' if x['Home Score'] > x['Away Score'] else 'Away', axis=1)
win_counts = match['Winner'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(win_counts, labels=win_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Home Wins vs. Away Wins')
plt.axis('equal')
plt.show()
```

```python
df_melted = match.melt(id_vars=['Home', 'Away', 'Date'], 
                       value_vars=['Home Score', 'Away Score'],
                       var_name='Score Type', 
                       value_name='Score')

plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x='Home', y='Score', hue='Score Type', palette='pastel')
plt.xticks(rotation=45, ha='right')
plt.title('Home vs Away Score for Each Team')
plt.xlabel('Match')
plt.ylabel('Score')
plt.tight_layout()
plt.show()
```
